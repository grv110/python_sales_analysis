```python
#import python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sales_data.csv', encoding='latin1')
df
```

```python
df.head()
```

```python
# fetch null data from amount col:
df[df['Amount'].isnull()]
```

```python
# drop unwanted col such as status and unnamed
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#Drop nan values
df.dropna(inplace=True)
df
```

```python
df.info()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

```python
df.head()
```

```python
# Rename coloum name
df.rename(columns={'Marital_Status':'Marital'},inplace=True)
```

```python
df.head()
```

```python
# change the martital satatus 0:singel,1": married
df['Marital']=df['Marital'].map({0:'Single',1:'Married'})
```

```python
df.head()
```

```python
df['Gender']=df['Gender'].map({"F":'Female',"M":'Male'})
df.head()
```

**Gender**

```python
a=sns.countplot(x='Gender',data=df,hue="Gender")
for i in a.containers:
  a.bar_label(i)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

**Age**

```python
a = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in a.containers:
    a.bar_label(bars)




```

```python
a=df.groupby('Gender')['Amount'].sum()
a=pd.DataFrame(a)

# visulize male and female by amount:
b=sns.barplot(x='Gender',y='Amount',data=a,hue='Gender',palette='Set1',width=0.5)
for i in b.containers:
  b.bar_label(i)
```

```python
def bar_count(a):
  for i in a.containers:
    a.bar_label(i)
```

**Count by age group**

```python
# count by age group
plt.figure(figsize=(10,5))
a=sns.countplot(x='Age Group',data=df,hue='Age Group',palette='bright',width=0.5)
bar_count(a)
plt.title('count by age group')
plt.show()
```

**Customer who are from 26 to 35 age group are more compared to other age group**

```python
plt.figure(figsize=(15,5))
Age_Group =df.groupby('Age Group')['Amount'].sum()
Age_Group=pd.DataFrame(Age_Group).reset_index()
s=sns.barplot(x='Age Group',y='Amount',data=Age_Group,hue='Age Group',palette='dark',width=0.5)
bar_count(s)
plt.title('Total amount by age group')
plt.show()
```

**Total amount by Occupation**

```python
plt.figure(figsize=(15,5))
Age_Group =df.groupby('Occupation')['Amount'].sum()
Age_Group =pd.DataFrame(Age_Group).reset_index()
s=sns.barplot(x='Occupation',y='Amount',data=Age_Group,hue='Occupation',palette='bright',width=0.7)
bar_count(s)
plt.title('Total amount by Occupation')
plt.xticks(rotation=45)
plt.show()
```

**From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector**

**Total amount by State**

```python
plt.figure(figsize=(20,5))
Age_Group=df.groupby('State')['Amount'].sum()
Age_Group=pd.DataFrame(Age_Group).reset_index()
s=sns.barplot(x='State',y='Amount',data=Age_Group,hue='State',palette='bright',width=0.7)
plt.title('Total amount by Statewise')
plt.xticks(rotation=45)
plt.show()
```
